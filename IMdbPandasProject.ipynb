```python
import pandas as pd
```

```python
movies = pd.read_csv(r"C:\Users\parml\Desktop\AB\Projects\imdb_movies.csv")
```

```python
movies.head()
```

```python
movies.info()
```

```python
pd.set_option('display.max.rows', 11000) # Show full table
pd.set_option('display.max.columns', 22)
```

```python
movies.head()
```

```python
##Check duplicate
movies[movies.duplicated()]
movies.drop_duplicates(inplace = True)
```

```python
movies[movies.duplicated()]
```

```python
#generes exploration so no null in genre--see Info output
movies.dropna(subset = ['genres'], inplace = True)
```

```python
##profit-make column Rev-Budget
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
#get rid of extra columns for memory and organization
movies_genre = movies[['popularity', 'budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_year', 'vote_count', 'vote_average', 'profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame ##important
```

```python
## Important b/c assigns all the genres to the index numb
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'  #created column with indiv generes
del movies_genre['genres']
movies_genre = movies_genre.join(split)
```

```python
movies_genre
```

```python
#1. Which genres are the most common(number of movies made?)
genres_count=pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = True)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title = "Movies Per Genre in %", autopct= '%1.1f%%', figsize = (10,10))
```

```python
genres_count['original_title'].plot.barh(title = "Movies per Genre", color = 'DarkBlue', figsize = (10,9))
```

```python
##2. Which genres have the highest average budget and revenue?
genres_avg = movies_genre.groupby('genres_split').mean('numeric_only')
genres_avg
```

```python
pd.options.display.float_format = '{:2f}'.format
genres_avg
```

```python
genres_avg.sort_values('budget', ascending = True, inplace = True)
```

```python
genres_avg[['budget', 'revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('DarkBlue', 'c'), figsize = (10,9))
```

```python
##3. Which genres have the highest average profit?
genres_avg.sort_values('profit', ascending = True, inplace = True)
```

```python
genres_avg[['profit']].plot.barh(title = 'Profit by Genre', color = ('DarkBlue'), figsize = (10,9))
```

```python
##4. Which genres have the highest average popularity?
```

```python
genres_avg.sort_values('popularity', ascending = True, inplace = True)
```

```python
genres_avg[['popularity']].plot.barh(title = 'Popularity by Genre', color = ('DarkBlue'), figsize = (10,9))
```

```python
##5. Which genres have highest number of movies with a votion avg >= 8?
movies_genre.head()
```

```python
vote_fifty = movies_genre[(movies_genre['vote_count'] >= 50) & (movies_genre['vote_average'] >= 8)]
vote_zero = movies_genre[movies_genre['vote_average'] >= 8] 
```

```python
genres_vote1=pd.DataFrame(vote_fifty.groupby('genres_split').vote_average.nunique()).sort_values('vote_average', ascending = True)
```

```python
genres_vote1
```

```python
genres_vote1[['vote_average']].plot.barh(title = 'Vote_average by Genre', color = ('DarkBlue'), figsize = (10,9))
```

```python
genres_vote2=pd.DataFrame(vote_zero.groupby('genres_split').vote_average.nunique()).sort_values('vote_average', ascending = True)
```

```python
genres_vote2
```

```python
genres_vote2[['vote_average']].plot.barh(title = 'Vote_average by Genre', color = ('DarkBlue'), figsize = (10,9))
```

```python
final_genre = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize': (15,10)})
sns.heatmap(final_genre, cmap = 'coolwarm', linewidths = .5)
plt.title('Genres by Profit per Year')
```
